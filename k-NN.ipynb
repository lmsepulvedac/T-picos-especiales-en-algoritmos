# Clasificador de *k* vecinos más cercanos (*k*-NN)

Dado un conjunto de observaciones, el algoritmo de *k*-NN se basa en la medida de distancia, que está dada por:

$$d(X_i,X_j)=\sqrt{\sum_{l=1}^{p}(x_{li}-x_{lj})},$$

donde $X_i = [x_1i,x_2i, \ldots x_pi] \in \mathbf{X}$, $X_j = [x_1j,x_2j, \ldots x_pj] \in \mathbf{X}$, y $p$ el número de características. Una vez encontrados los *k* vecinos más cercanos, se utiliza una regla mayoritaria para determinar la etiqueta $c \in C$ para la nueva muestra.


```python
#importamos librerías
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn import neighbors, datasets
```

```python
archivo = 'data_knn.txt'
data = np.genfromtxt(archivo, delimiter = ',', skip_header = True) #cargamos los datos
X, y = data[:,:-1], data[:,-1].astype(np.int) #se asegura que la etiqueta sea de tipo entero
```

```python
#visualizamos los datos
plt.figure()
plt.title('Datos de entrada')
marcas = 'v^os'
marcas_expansion = [marcas[i] for i in y]
for i in range(X.shape[0]):
    plt.scatter(X[i,0],X[i,1], marker = marcas_expansion[i],
               s = 75, edgecolors = 'black', 
                facecolors = 'none')
plt.show()
    
```

```python
#definimos el número de vecinos
NN = 11
#instanciamos el clasificador
clasificador = neighbors.KNeighborsClassifier(NN, 
                                              weights = 'distance')
#entrenamos el clasificador
clasificador.fit(X,y)
```

```python
#validamos el clasificador
test = [5.1, 3.6]
y_pred = clasificador.predict([test])
print(y_pred)
```
